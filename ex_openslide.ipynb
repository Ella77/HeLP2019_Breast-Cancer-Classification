```python
import openslide
import numpy as np
import matplotlib.pyplot as plt
```

```python
slide = openslide.OpenSlide('./level0/Example_002.mrxs')
print(slide)
```

```python
for i, d in enumerate(slide.level_dimensions):
    print('resolution of level{} slide :'.format(i), d)
```

```python
# whole slide for level4
plt.imshow(slide.read_region([0, 0], 4, slide.level_dimensions[4]))
```

```python
plt.figure(figsize=(30,30))
for i in range(10):
    plt.subplot(1,5,(i%5)+1)
    plt.imshow(slide.read_region([20000, 44000], i, [512, 512]))
    plt.title('level{}'.format(i))
    if (i+1) % 5 == 0:
        plt.show()
        plt.figure(figsize=(30,30))
plt.show()
```
